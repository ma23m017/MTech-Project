## Question 2

How is the frequency of orders varying across weekdays, hour of day?

## Solution:

### Frequency of orders

```python
# Frequency of orders by day of the week
orders_by_dow = orders['order_dow'].value_counts().sort_index()
print('Orders by day of week is given by:')
print(orders_by_dow)

# Frequency of orders by hour of day
orders_by_hour = orders['order_hour_of_day'].value_counts().sort_index()
print('\nOrders by hour is given by:')
print(orders_by_hour)
```

```python
# Plotting
plt.figure(figsize=(14, 5))

# Plot for days of the week
plt.subplot(1, 2, 1)
sns.barplot(x=orders_by_dow.index, y=orders_by_dow.values)
plt.title('Orders by Day of the Week')
plt.xlabel('Day of the Week (0 = Sunday)')
plt.ylabel('Number of Orders')

# Plot for hours of the day
plt.subplot(1, 2, 2)
sns.barplot(x=orders_by_hour.index, y=orders_by_hour.values)
plt.title('Orders by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Orders')

plt.tight_layout()
plt.show()  # Displays the plot
```

From above plots we can say that most orders are placed on sunday and monday in terms of days and if we talk about hours of the day then people placed order between 9 am to 10 pm.

## Question 5

Identify top product categories/departments, categories with higher reorder rate.

## Solution:

### Reorder rate by department

```python
# Merge order data with products and departments
order_products = order_products.merge(products, on='product_id').merge(departments, on='department_id')
order_products
```

```python
# Calculate total orders and reorder rates per department
department_stats = order_products.groupby('department')['reordered'].agg(
    total_orders='size',
    reorder_rate='mean'
).reset_index()

# Sort by total orders and reorder rate
top_departments = department_stats.sort_values(by='total_orders', ascending=False)
top_departments_total_orders = top_departments.drop(columns=['reorder_rate'], inplace=False)

# Print top departments by total orders
print("Top Departments by Total Orders:")
top_departments_total_orders
#top_departments
```

```python
top_reorder_departments = department_stats.sort_values(by='reorder_rate', ascending=False) # Sorting by reorder rate
top_departments_reorder = top_reorder_departments.drop(columns=['total_orders'], inplace=False)

# Print top departments by reorder rate
print("Top Departments by Reorder Rate:")
top_departments_reorder
```

## Question 6

Which products are added to the cart first? Explore relationship between the order in which products are added to the cart and the reorder frequency.

## Solution:

```python
# Combining prior order and train order table
order_products = pd.concat([order_products_prior, order_products_train])

# Selecting products which are added to cart first
first_products = order_products[order_products['add_to_cart_order'] == 1]['product_id'].value_counts().head(10)
first_products = first_products.reset_index()

first_products.columns = ['product_id', 'count']
first_products
```

```python
# Merging first_product and products table on product_id
first_products = pd.merge(first_products, products, on='product_id', how='left')

# Printing the top 10 products which are added to the cart first
print('Top 10 products which are added to the cart first are:\n')
first_products[['product_id', 'product_name', 'count']]
```

### Relationship

```python
# Merge with products to get product names
order_products = order_products.merge(products[['product_id', 'product_name']], on='product_id', how='left')

# Calculate reorder frequencies by add-to-cart order
reorder_freq = order_products.groupby('add_to_cart_order')['reordered'].mean().reset_index()

# Plotting the relationship between add-to-cart order and reorder frequency
plt.figure(figsize=(12, 6))
sns.lineplot(x='add_to_cart_order', y='reordered', data=reorder_freq)
plt.title('Reorder Frequency vs Add to Cart Order')
plt.xlabel('Add to Cart Order')
plt.ylabel('Reorder Frequency')

plt.grid(True)
plt.tight_layout()
plt.show() # Displaying the plot
```

## Question 7

How is the variation in the basket size of customers across the history of prior orders?

## Solution:

```python
# Load the dataset
orders = pd.read_csv('orders.csv')
order_products_prior = pd.read_csv('order_products__prior.csv')

# Calculate the Basket Size for Each Order
basket_size_per_order = order_products_prior.groupby('order_id')['product_id'].count().reset_index()
basket_size_per_order.columns = ['order_id', 'basket_size']

# Merge Basket Size with Orders to Get User IDs
orders_with_basket_size = orders.merge(basket_size_per_order, on='order_id', how='left')

# Filter only prior orders
prior_orders = orders_with_basket_size[orders_with_basket_size['eval_set'] == 'prior']

# Calculate Basket Size Statistics for Each Customer
customer_basket_stats = prior_orders.groupby('user_id')['basket_size'].agg(['mean', 'std', 'min', 'max']).reset_index()
customer_basket_stats.columns = ['user_id', 'mean_basket_size', 'std_basket_size', 'min_basket_size', 'max_basket_size']

# Display the first 10 rows of the table
print(customer_basket_stats.head(10))

# Optionally, save the table to a CSV file
customer_basket_stats.to_csv('customer_basket_stats.csv', index=False)
```

## Question 8

Do customers with higher no. of orders have higher proportion of reordered products?

## Solution:

```python
# Again loading the dataset
orders = pd.read_csv('orders.csv')
order_products_prior = pd.read_csv('order_products__prior.csv')

# Calculate the Total Number of Orders for Each Customer
prior_orders = orders[orders['eval_set'] == 'prior']
total_orders_per_customer = prior_orders.groupby('user_id')['order_id'].nunique().reset_index()
total_orders_per_customer.columns = ['user_id', 'total_orders']
```

```python
# Calculate the Number of Reordered Products for Each Customer
order_products_prior = order_products_prior.merge(prior_orders[['order_id', 'user_id']], on='order_id', how='left')
total_products_per_customer = order_products_prior.groupby('user_id')['product_id'].count().reset_index()
total_products_per_customer.columns = ['user_id', 'total_products']

reordered_products_per_customer = order_products_prior[order_products_prior['reordered'] == 1].groupby('user_id')['product_id'].count().reset_index()
reordered_products_per_customer.columns = ['user_id', 'reordered_products']
```

```python
# Calculate the Proportion of Reordered Products for Each Customer
customer_reorder_data = total_products_per_customer.merge(reordered_products_per_customer, on='user_id', how='left')
customer_reorder_data = customer_reorder_data.merge(total_orders_per_customer, on='user_id', how='left')
customer_reorder_data['reorder_proportion'] = customer_reorder_data['reordered_products'] / customer_reorder_data['total_products']

# Showing 10 reorder data
customer_reorder_data = customer_reorder_data.sort_values(by='reorder_proportion', ascending=False).head(10)
customer_reorder_data
```

```python
# Calculating correlation to find the relation between total orders and reorder proportion
correlation = customer_reorder_data[['total_orders', 'reorder_proportion']].corr().iloc[0, 1]
print(f'Correlation between total orders and reorder proportion: {correlation}')
```

From the correlation we cannot say that customers with higher no. of orders have higher proportion of reordered products.
